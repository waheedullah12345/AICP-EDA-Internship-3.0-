Q.2: **Check statistical facts by checking all columns. Then calculate the maximum value of the following
attributes ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (%
Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)']**

```python
# Check statistical facts of all columns
statistical_facts = df.describe()

# Calculate the maximum value of the specified attributes
max_values = df[['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein',
                 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)',
                 'Iron (% Daily Value)']].max()

statistical_facts, max_values
```

Q.3: **Check to see if infact there is any correlation between Calories and other independent variables by
plotting a correlation matrix next.**

```python
# Assuming 'df' is your DataFrame after reading the CSV file

# Selecting the columns of interest
columns_of_interest = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars',
                       'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)',
                       'Calcium (% Daily Value)', 'Iron (% Daily Value)']

# Calculating the correlation matrix for the selected columns
correlation_matrix = df[columns_of_interest].corr()

# Plotting the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

Q.4: **Draw boxplot for Calories vs Category to spot outliers and max calories category**

```python
# Create a boxplot for Calories vs Category
plt.figure(figsize=(14, 8))
sns.boxplot(x='Calories', y='Category', data=df, palette="Set3")
plt.title('Calories by Category')
plt.show()

# Identify the category with the maximum calories
max_calories_category = df.loc[df['Calories'].idxmax(), 'Category']
print(f"The category with the maximum calories is: {max_calories_category}")
```

Q.6: **Draw Stripplot for each
category against the following
attributes ['Calories', 'Total Fat',
'Carbohydrates', 'Dietary
Fiber', 'Sugars', 'Protein',
'Vitamin A (% Daily Value)',
'Vitamin C (% Daily Value)',
'Calcium (% Daily Value)', 'Iron
(% Daily Value)'.**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('/content/menu (1).csv')

# Set the style of seaborn to 'ticks' for a cleaner look with ticks
sns.set(style="ticks")

# Create the strip plot with a specified palette
# Here 'deep' is a seaborn palette. You can also use 'muted', 'bright', 'pastel', 'dark', or 'colorblind'.
plt.figure(figsize=(12, 8))
strip_plot = sns.stripplot(x="Category", y="Calories", data=df, jitter=True, palette='deep')

# Rotate the category labels to prevent overlap
strip_plot.set_xticklabels(strip_plot.get_xticklabels(), rotation=45)

# Add a title and adjust the axis labels as necessary
plt.title('Calories Content by Category')
plt.xlabel('Category')
plt.ylabel('Calories')

# Adjust the layout to make room for the rotated category labels
plt.tight_layout()

# Show the plot
plt.show()
```

Q.7: **Draw a horizontal bar graph for items in each category against the calories.
Here is one sample for all items in Beef & Pork, you have to do it for items in each category. Also, write
your observation.**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/menu (1).csv')

# Filter the DataFrame for the 'Beef & Pork' category
beef_pork_df = df[df['Category'] == 'Beef & Pork']

# Sort the items by calories in ascending order so that the lowest calorie items are at the bottom
beef_pork_df_sorted = beef_pork_df.sort_values('Calories', ascending=True)

# Draw a horizontal bar graph
plt.figure(figsize=(10, 8))
plt.barh(beef_pork_df_sorted['Item'], beef_pork_df_sorted['Calories'], color='skyblue')
plt.xlabel('Calories')
plt.title('Calorie Content of Beef & Pork Items (Low to High)')
plt.tight_layout()  # Adjust the plot to ensure everything fits without overlapping
plt.show()
```

**Observations from the graph:**

1. The **Double Quarter Pounder with Cheese** has the highest calorie content in this category, which is significantly higher than the other items.
2. The **Hamburger** has the lowest calorie content, making it the least calorific option in this category.
3. There's a noticeable incremental increase in calorie content from simpler items like the **Hamburger** and **Cheeseburger** to more complex ones like the **Big Mac** and the **Quarter Pounder** variants.
4. Most of the items in this category have a high-calorie count, often exceeding 500 calories, which is substantial considering these are individual items and not full meals.
